My first project of functions: The last prime number in a range

```python
def is_prime (n):
    prime = True
    for i in range(2,int(n**.5)+1): # [ (n**.5)+1 ] this action is done to reduce the running time.
        if n % i == 0:
            prime = False
            break
    return (prime)

the_last_prime_number = 0
for i in range (2, 101):
    if is_prime((i)):
        the_last_prime_number = i
    if is_prime((i)):
        print(i , is_prime (i))
print('the_last_prime_number: ' , the_last_prime_number)
```
