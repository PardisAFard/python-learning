Learning how to use numpy and matplotlib libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace

x = linspace(0, 10, 100)

fig , axs = plt.subplots(2, 2 , figsize= (8,4)) # creating a matrix with four arrays.

axs[0, 0]. plot (x, np.sin(x), color= 'red')
axs[0, 0]. set_title('Sin(x)')

axs[0, 1]. plot (x, np.cos(x), color= 'g')
axs[0, 1]. set_title('Cos(x)')

axs[1, 0]. plot (x, np.tan(x), color= 'blue')
axs[1, 0]. set_title('tan(x)')

axs[1, 1]. plot (x, np.exp(x), color= 'yellow')
axs[1, 1]. set_title('e^(x)')

plt. tight_layout() # creating a good space between the plots.
plt.show()
```
