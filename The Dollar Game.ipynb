The Dollar Game – A Wealth Inequality Simulator

```python
import random
random.seed 
people = []
for i in range(0 , 50):
    people.append(100)
#print(people)
for i in range(0 , 10000):
    for p1 in range(0 , 50):
        p2 = random.randrange(0 , 50) # choose the reciving person.
        while people[p2] == 0: # if the reciving person has 0 money, choose another one.
            p2 = random.randrange(0 , 50)
        if people[p1] != 0:
            people[p1] = people[p1] - 1
            people[p2] = people[p2] + 1
        #print(p1 , p2)
print(people)

import matplotlib.pyplot as plt
people.sort(reverse = True)
#people.sort()
plt.bar(range(0 , 50), people) 
plt.ylabel('reciving money')
plt.xlabel('people') 
```
